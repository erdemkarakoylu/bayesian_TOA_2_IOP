```python
abs_bands = [411, 443, 489, 510]
sat_bands = [412, 443, 490, 510, 555, 670]
env_columns = 'lat,lon,id,oisst,etopo2,cruise'.split(',')
target_cols = 'chl,chl_a'.split(',') 
target_cols += [f'ad{ab}' for ab in abs_bands]
target_cols += [f'ap{ab}' for ab in abs_bands]
rrs_columns = [f'sat_rrs{b}' for b in sat_bands]
toa_columns = [
    f'sat_lt{b}' for b in sat_bands] + [f'sat_rhot{b}' for b in sat_bands]
sensor_cols = 'solz,sola,senz,sena'.split(',')
```

```python
df12_env = df12[env_columns]
df12_targets = df12[target_cols]
df12_rrs = df12[rrs_columns]
df12_toa = df12[toa_columns]
```

```python
save_12_path = data_path / 'reduced_columns' / 'nomad_2008_2012'
df12_env.to_parquet(save_12_path/'df_env.pqt')
df12_targets.to_parquet(save_12_path / 'df_targets.pqt')
df12_rrs.to_parquet(save_12_path / 'df_rrs.pqt')
df12_toa.to_parquet(save_12_path / 'df_toa.pqt')
```
